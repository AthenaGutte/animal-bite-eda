```python
bites['bite_date'] = pd.to_datetime(bites['bite_date'], errors='coerce')
bites['bite_year'] = bites['bite_date'].dt.year
bites['bite_month'] = bites['bite_date'].dt.month
bites['bite_date'].isna().sum()
```

Handling Null Values

```python
bites.isna().sum()
```

```python
categorical_cols=['SpeciesIDDesc', 'BreedIDDesc', 'GenderIDDesc', 'color', 'AdvIssuedYNDesc', 'WhereBittenIDDesc']
bites[categorical_cols] = bites[categorical_cols].fillna('UNKNOWN')
bites['victim_zip'] = bites['victim_zip'].fillna('UNKNOWN')
bites['vaccination_yrs'] = bites['vaccination_yrs'].fillna(bites['vaccination_yrs'].median())
bites['bite_year'] = bites['bite_year'].fillna(bites['bite_year'].median())
bites['bite_month'] = bites['bite_month'].fillna(bites['bite_month'].median())
bites
```

```python
bites.isna().sum()
```

```python
bites['bite_year'].value_counts().sort_index()
```

```python
bites = bites[(bites['bite_year'] >= 1952) & (bites['bite_year'] <= 2021)]
```
